```python
import numpy as np
import matplotlib.pyplot as plt

# input data
X = np.array([[0,0],[0,1],[1,0],[1,1]])
# output data
y = np.array([0,0,0,1])

# weights
w = np.array([0.0,0.0])
# bias
b = 0.0

# learning rate
alpha = 0.1

# activation function
def activation(x):
    if x >= 0:
        return 1
    else:
        return 0
    
# prediction function
def predict(X):
    return activation(np.dot(X,w) + b)

# training function
def train(X,y,b):
    for _ in range(10):
        for i in range(len(X)):
            pred = predict(X[i])
            w[0] += alpha * (y[i] - pred) * X[i][0]
            w[1] += alpha * (y[i] - pred) * X[i][1]
            b += alpha * (y[i] - pred)

# training
train(X,y,b)

# testing
for i in range(len(X)):
    print("X: {} y: {} pred: {}".format(X[i], y[i], predict(X[i])))

# plotting
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AND')
plt.legend(['0','1'])
plt.show()
```

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('horseRacing.csv')
data.head()
```

```python
data.describe()
```

```python
# drop last column
# convert labels to binary
# visualize the skewness of the data
# drop rows without any features, only first 2 and last 2 columns are present. Drop these complete raceids
# if a single row follows this pattern than drop all rows with that raceid
# impute missing values with mean
# change both ids to one hot encoding
```

```python
# drop last column of price

data = data.drop('price', axis=1)
data.head()
```

```python
# convert calc_position to binary

data['calc_position'] = data['calc_position'].apply(lambda x: 1 if x >= 3 else 0)
data.describe()
```

```python
# visualize the skewness of the data

data.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# drop rows without any features, only first 2 and last 2 columns are present. Drop these complete raceids

data = data.dropna(subset=['race_id','horse_id', 'calc_position'])
data.describe()
```

```python
# impute missing values with mean of that column

data = data.fillna(data.mean())
data.describe()
```

```python
data.head()
```

```python
# normalize data using standard scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

for col in data.columns[2:-2]:
    data[col] = scaler.fit_transform(data[[col]])

data.head()
```

```python
temp = data['race_id']
temp1 = data['horse_id']

temp = pd.get_dummies(temp)
temp1 = pd.get_dummies(temp1)
```

```python
# applying MLP classifier on the data
from sklearn.neural_network import MLPClassifier

# split data into train and test
from sklearn.model_selection import train_test_split

X = data.drop(['race_id','horse_id', 'calc_position'], axis=1)
y = data['calc_position']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,random_state=1)
clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
clf.predict_proba(X_test)
clf.predict(X_test)
clf.score(X_test, y_test)
```
